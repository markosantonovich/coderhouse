```python
import google.generativeai as genai

API_KEY = "API_KEY(no la pongo por las dudas jeje)"
genai.configure(api_key=API_KEY)
```

```python
def generar_prompt(ocupacion, dificultad, objetivo):

    #Genera un prompt personalizado seg√∫n el perfil del usuario.

    return f"Explicame c√≥mo {objetivo} para alguien que trabaja de {ocupacion} y tiene dificultad con {dificultad}."
```

```python


def obtener_respuesta(prompt):
    modelo = genai.GenerativeModel("models/gemini-2.5-flash-preview-05-20")
    respuesta = modelo.generate_content(prompt)
    return respuesta.text
```

```python
# Datos simulados del usuario
ocupacion = "t√©cnico en refrigeraci√≥n"
dificultad = "usar Excel"
objetivo = "crear una hoja de c√°lculo para llevar los gastos del taller"

# Generar prompt personalizado
prompt = generar_prompt(ocupacion, dificultad, objetivo)
print("‚úÖ Prompt generado:")
print(prompt)

# Obtener respuesta desde Gemini 2.5 Flash
respuesta = obtener_respuesta(prompt)
print("\nüí¨ Respuesta de la IA:")
print(respuesta)
```
